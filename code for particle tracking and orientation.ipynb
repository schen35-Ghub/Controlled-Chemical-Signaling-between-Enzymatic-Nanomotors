```python
'''
EXPLANATION OF SCRIPT 1: BACKGROUND SUBTRACTION AND NOISE REDUCTION USING GUI

The goal of this first script is to extract the background from single-particle videos 
containing nanomotors (NMs). To facilitate this process, a graphical user interface (GUI) 
is provided to make it more user-friendly and interactive.

The background extraction process follows these steps:

    1. A set of 100 random frames is selected from the original video.
    2. For each pixel location, the median intensity across these frames is computed.
       This median image represents the static background.
    3. Each frame in the video is then subtracted from the background using 
       the absolute difference to highlight moving objects (NMs).
    4. A threshold is applied to filter out low-intensity noise and enhance 
       contrast between NMs and background.
    5. Finally, the result is improved using a blur (smoothing) method, which
       helps to reduce small noise and fill gaps in segmented regions.
       
'''

'''
THINGS TO DO: 

ŌåÆ Run the code and load a video using the GUI.
ŌåÆ Adjust the parameters as needed to improve NMs detection.
ŌåÆ Click "Preview First Frames" to preview the result of the processing on the first 50 frames.
ŌåÆ If you're not satisfied with the preview, change the parameter values 
   and click "Preview First Frames" again. Repeat as needed until the result is acceptable.
ŌåÆ Once satisfied with the result, click "Process Full Video" 
   to apply the processing to the full dataset and save the output.

Threshold (NR):
- Controls the sensitivity in distinguishing NMs from the background.  
- Lower values allow more noise but capture more NMs; higher values exclude weak signals.  
    * For videos with strong contrast between NMs and the background, try lower values (e.g., 10).  
    * For low-contrast videos, use higher values.

Kernel Size (ksize):
- Defines the size of the area used in the blur filter.  
- Larger kernels smooth more but may blur fine details.  
    * Typical values: (2,2), (3,3), (4,4), or (5,5)

Iterations Blur (Number of Blurs):
- Determines how many times the blur is applied.  
- More iterations improve noise reduction but can oversmooth features.  
    * Typical values: 1 (most common), 2, or 3.

'''

from __future__ import division, unicode_literals, print_function
import cv2
import numpy as np
import os
import tkinter as tk
from tkinter import filedialog, messagebox
from PIL import Image, ImageTk
from tkinter.ttk import Progressbar  # no se usar├Ī pero puedes dejarla si quieres
import time
import threading
import sys
import pims

# ============ Global Variables ============
file_name = None
file_name_input = None

# ============ Video Processing ============
def process_video(video_path, NR, ksize, num_blurs, max_frames=None):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error opening the video")
        return None

    num_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(3))
    height = int(cap.get(4))

    frame_count = min(100, num_frames)
    frame_ids = np.random.uniform(size=frame_count) * num_frames
    frames = []

    for fid in frame_ids:
        cap.set(cv2.CAP_PROP_POS_FRAMES, fid)
        ret, frame = cap.read()
        if not ret:
            continue
        frames.append(frame)

    median = np.median(frames, axis=0).astype(np.uint8)
    median = cv2.cvtColor(median, cv2.COLOR_BGR2GRAY)

    cap.set(cv2.CAP_PROP_POS_FRAMES, 0)

    output_frames = []
    frame_idx = 0

    while True:
        if max_frames and frame_idx >= max_frames:
            break

        ret, frame = cap.read()
        if not ret:
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        dif_frame = cv2.absdiff(median, gray)

        for _ in range(num_blurs):
            dif_frame = cv2.blur(dif_frame, ksize)

        _, result = cv2.threshold(dif_frame, NR, 255, cv2.THRESH_BINARY_INV)
        output_frames.append(result)
        frame_idx += 1

    cap.release()
    return output_frames, fps, (width, height)

# ============ Save Full Video ============
def save_video(frames, fps, size, output_name):
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_name, fourcc, fps, size, isColor=False)

    for frame in frames:
        out.write(frame)

    out.release()

# ============ Additional Processing (frames y trackpy) ============
def run_additional_process():
    global file_name

    import matplotlib as mpl
    import matplotlib.pyplot as plt
    mpl.rc('image', cmap='gray')
    import pandas as pd
    import trackpy as tp
    from PIL import Image
    import warnings
    warnings.filterwarnings("ignore", category=FutureWarning)

    my_id = '01'
    my_folder = f'frames_{my_id}'
    os.makedirs(my_folder, exist_ok=True)
    stream = cv2.VideoCapture(file_name)
    if not stream.isOpened():
        print("No video")
        exit()
    num_frames = int(stream.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = stream.get(cv2.CAP_PROP_FPS)
    width = int(stream.get(3))
    height = int(stream.get(4))
    stream.set(cv2.CAP_PROP_POS_FRAMES, 0)
    for i in range(num_frames):
        ret, frame = stream.read()
        if not ret:
            break
        frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        im = Image.fromarray(frame_gray)
        frame_name = f"{my_folder}/frame{i:04d}.png"
        im.save(frame_name)
    stream.release()
    cv2.destroyAllWindows()
    frames = pims.ImageSequence(f'frames_{my_id}/*.png')

# ============ GUI Class ============
class VideoGUI:
    def __init__(self, root):
        self.root = root
        self.root.title("Background Extractor - GUI")

        self.video_path = None
        self.file_name_input = None
        global file_name, file_name_input

        self.nr_var = tk.IntVar(value=10)
        self.ksize_x = tk.IntVar(value=3)
        self.ksize_y = tk.IntVar(value=3)
        self.num_blurs = tk.IntVar(value=1)

        tk.Button(root, text="Select Video", command=self.select_video).pack(pady=5)

        self.info_label = tk.Label(root, text="No video selected", fg="red")
        self.info_label.pack()

        tk.Label(root, text="NR (background threshold)").pack()
        tk.Scale(root, from_=1, to=100, orient="horizontal", variable=self.nr_var, length=500).pack()

        tk.Label(root, text="Kernel size (x, y)").pack()
        frame_k = tk.Frame(root)
        frame_k.pack()
        tk.Entry(frame_k, textvariable=self.ksize_x, width=5).pack(side="left")
        tk.Entry(frame_k, textvariable=self.ksize_y, width=5).pack(side="left")

        tk.Label(root, text="Number of blurs").pack()
        tk.Scale(root, from_=1, to=7, orient="horizontal", variable=self.num_blurs).pack()

        tk.Button(root, text="Preview First Frames", command=self.preview_analysis).pack(pady=10)
        tk.Button(root, text="Process Full Video", command=self.full_analysis).pack(pady=10)

        self.canvas = tk.Label(root)
        self.canvas.pack()

    def select_video(self):
        path = filedialog.askopenfilename(filetypes=[("Video files", "*.mp4;*.avi;*.mov;*.mkv")])
        if path:
            self.video_path = path
            self.file_name_input = os.path.basename(path)
            global file_name_input
            file_name_input = self.file_name_input
            self.info_label.config(text=f"Video: {self.file_name_input}", fg="green")

    def preview_analysis(self):
        if not self.video_path:
            messagebox.showerror("Error", "Please select a video first.")
            return

        NR = self.nr_var.get()
        ksize = (self.ksize_x.get(), self.ksize_y.get())
        blurs = self.num_blurs.get()

        frames, _, _ = process_video(self.video_path, NR, ksize, blurs, max_frames=50)

        cap = cv2.VideoCapture(self.video_path)
        width = int(cap.get(3))
        height = int(cap.get(4))

        n  =  int(width/512)

        self.frames_resized = [cv2.resize(f, (int(width / n), int(height / n))) for f in frames]

        
        self.frames_gui = [ImageTk.PhotoImage(image=Image.fromarray(f)) for f in self.frames_resized]
        self.current_frame = 0
        self.play_frames()

    def play_frames(self):
        if hasattr(self, "frames_gui") and self.frames_gui:
            self.canvas.config(image=self.frames_gui[self.current_frame])
            self.current_frame = (self.current_frame + 1) % len(self.frames_gui)
            self.root.after(50, self.play_frames)

    def full_analysis(self):
        if not self.video_path:
            messagebox.showerror("Error", "Please select a video first.")
            return

        NR = self.nr_var.get()
        ksize = (self.ksize_x.get(), self.ksize_y.get())
        blurs = self.num_blurs.get()

        video_path = self.video_path
        file_base = os.path.splitext(self.file_name_input)[0]
        global file_name
        file_name = f"{file_base}_no_background.mp4"

        self.root.destroy()

        # Terminal animation
        stop_event = threading.Event()

        def animation():
            symbols = ['0oo', 'o0o', 'oo0']
            i = 0
            while not stop_event.is_set():
                sys.stdout.write(f'\rProcessing... {symbols[i % 3]}')
                sys.stdout.flush()
                i += 1
                time.sleep(0.3)

        t = threading.Thread(target=animation)
        t.start()

        frames, fps, size = process_video(video_path, NR, ksize, blurs)
        save_video(frames, fps, size, file_name)
        run_additional_process()

        stop_event.set()
        t.join()

        sys.stdout.write('\r' + ' ' * 30 + '\r')  # Borra la l├Łnea actual en la consola
        sys.stdout.flush()
        print(f"Done! Video saved as: {file_name}")


# ============ Run GUI ============
if __name__ == "__main__":
    root = tk.Tk()
    app = VideoGUI(root)
    root.mainloop()
```

```python
'''
EXPLANATION OF SCRIPT 3: PARTICLE SIZE AND SHAPE FILTERING IN A SINGLE FRAME

The goal of this third script is to analyze the size and shape of particles (NMs)
within a single selected frame from the background-subtracted video.

It produces both a visual and quantitative summary of particle size distribution by grouping 
particles into defined diameter ranges (in pixels). Additionally, it filters particles based on 
two criteria:
    - max_diameter: the maximum allowed particle diameter (in pixels).
    - max_eccentricity: a measure of particle elongation (0 = perfect circle, 1 = highly elongated).

This script is intended exclusively for inspection and parameter tuning. It does not modify the full dataset, 
but helps you understand particle size characteristics and choose appropriate filtering parameters 
before running the full tracking analysis.

'''

'''
THINGS TO DO:

ŌåÆ Adjust the following parameters to filter out unwanted particles:

selected_frame:  
- The index of the frame to analyze.  
- Choose any frame from 0 up to the total number of frames in the video.

max_diameter: 
- The maximum allowed particle diameter (in pixels).  
- Particles larger than this value will be excluded from the filtered image.  
  * Recommendation: Use the visual output and the particle size distribution table 
    to determine a suitable value.

max_eccentricity:  
- The maximum allowed elongation of a particle.  
- Ranges from 0 (perfect circle) to 1 (highly elongated).  
  * Helps filter out non-circular or irregularly shaped particles.  
  * Typical value: 0.85
  
'''

########################      
selected_frame = 750
max_diameter = 20
max_eccentricity = 0.85
########################  


import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import label, regionprops

binary = frames[selected_frame] > 0
binary_invertida = np.logical_not(binary)
labeled_inv = label(binary_invertida)
props_inv = regionprops(labeled_inv)

ranges = [(0, 3), (3, 6), (6, 10), (10, 15), (15, 20), (20, 25), (25, np.inf)]
counts = [0] * len(ranges)

filtered_inv = np.copy(binary_invertida)
for region in props_inv:
    diameter = region.equivalent_diameter
    for i, (low, high) in enumerate(ranges):
        if low < diameter <= high:
            counts[i] += 1
            break

            
    if (region.equivalent_diameter > max_diameter or 
        region.eccentricity > max_eccentricity):
        
        filtered_inv[labeled_inv == region.label] = 0


print("Distribution of particles by diameter range (in pixels):")
for (low, high), count in zip(ranges, counts):
    high_label = f"{int(high)}" if high != np.inf else "Ōł×"
    print(f"  {low:.0f}ŌĆō{high_label} px: {count} particles")

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(binary, cmap="gray")
axes[0].set_title("Original binarized")
axes[1].imshow(binary_invertida, cmap="gray")
axes[1].set_title("Original Binarized and inverted")
axes[2].imshow(filtered_inv, cmap="gray")
axes[2].set_title("Filtered: Diameter and Eccentricity")
for ax in axes:
    ax.axis("off")
plt.tight_layout()
plt.show()
```

```python
'''
EXPLANATION OF SCRIPT 4: FULL VIDEO PARTICLE SIZE AND SHAPE FILTRATION

The goal of this fourth script is to apply the particle size and shape filtering 
to ALL frames of the background-subtracted video.  
Using the thresholds defined earlier (`max_diameter` and `max_eccentricity`), 
the script removes any object that is too large or too elongated, ensuring that 
only the nanomotors of interest remain in the dataset used for Trackpy analysis.

'''

'''
THINGS TO DO:

ŌåÆ Set the parameters `max_diameter` and `max_eccentricity` as determined in the previous step [SCRIPT 3].  
ŌåÆ Run the script to apply the filtering across all frames.  

'''

#########################
max__diameter = 10
max__eccentricity = 0.85
#########################


import numpy as np
import cv2
from skimage.measure import label, regionprops
filtered_frames = []


for i, frame in enumerate(frames):
    binary = frame > 0  
    binary_inverted = np.logical_not(binary)
    labeled_frame = label(binary_inverted)
    props = regionprops(labeled_frame)
    filtered_binary = np.copy(binary_inverted)
    for region in props:
        if (region.equivalent_diameter > max__diameter 
            # ELIMINATE IF NECESSARY
           ):
            filtered_binary[labeled_frame == region.label] = 0  
    filtered_frames.append(filtered_binary)
    if i % 100 == 0:
        print(f"Processed {i}/{len(frames)} frames")
print("Process completed. All frames have been inverted and filtered.")
```

```python
'''
EXPLANATION OF SCRIPTŌĆ»5: DEFINING SPATIAL AND TEMPORAL PARAMETERS

The goal of this script is to define the spatial and temporal scales  
that will be used in all subsequent single-particle tracking analyses.

'''

'''
THINGS TO DO:

ratio_u_p:  
- Set the ratio of micrometers per pixel, based on your microscopeŌĆÖs calibration.

time_s:  
- Duration (in seconds) to study the Mean Squared Displacement (MSD), diffusion coefficient,  
  and velocity of the NMs.  
  * Typical values range from 1.5 to 2 seconds.


'''

#########################
ratio_u_p = 332.64/2048.   
time_s = 30                 
#########################

max_lagtime = int(fps * time_s)
```

```python
'''
EXPLANATION OF SCRIPTŌĆ»6: TRACKPY PARTICLE DETECTION INITIALIZATION

The goal of this script is to detect the positions of nanomotors (NMs) in every  
filtered frame, so that their trajectories can be constructed later.

'''

'''
THINGS TO DO:

ŌåÆ Adjust the two tuning parameters below and run the code:

diameter_num:  
- An odd-numbered estimate of particle diameter in pixels.  
- Use a value equal to or slightly less than the `max_diameter` set in ScriptŌĆ»4.  
  * Example: if `max_diameterŌĆ»=ŌĆ»25`, try `diameterŌĆ»ŌēłŌĆ»23`.

separation_num:  
- Minimum centreŌĆætoŌĆæcentre distance (in pixels) between two particles.  
- Must be ŌēźŌĆ»diameter. Larger values reduce overlapping detections, but may detect fewer particles.   
  * Recommendation: use 1ŌĆō2ŌĆ»├ŚŌĆ»diameter.

'''

####################
diameter_num = 11
separation_num = 20
####################

import pandas as pd, datetime
import trackpy as tp

f2 = tp.batch(filtered_frames, diameter=diameter_num, invert=False, separation=separation_num) 
```

```python
'''
EXPLANATION OF SCRIPTŌĆ»7: TRACKPY TRAJECTORY RECONSTRUCTION

The goal of this script is to convert detected points into complete particle trajectories..

'''

'''
THINGS TO DO:

max_displacement:  
- Set the maximum distance (in pixels) a particle can move between consecutive frames and still be considered the same particle.  
- Typical range: 10-25 (higher for faster-moving particles).


memory:  
- Specifies how many frames a particle can temporarily disappear and still be linked to the same trajectory.  
- The ideal value depends on the frame rate (fps); for example, at 40 fps, a memory between 8 and 18 frames is effective.  
- Increase this value if particle detections are frequently missed.

'''

######################
max_displacement = 20
memory_num = 15
######################

t = tp.link(f2, max_displacement, memory=memory_num)

t1 = tp.filter_stubs(t, max_lagtime) 

print('Before:', t['particle'].nunique())  
print('After:', t1['particle'].nunique())  

tp.plot_traj(t1)
```

```python
'''
EXPLANATION OF SCRIPT 8: FILTERING STATIC PARTICLES IN TRACKING DATA

The goal of this script is to (optionally) remove ŌĆ£stuckŌĆØ NMs from the
tracking dataset.

Particles that show zero movement for 5 or more frames are considered stuck and 
can be excluded from further analysis to improve the quality of tracking results.

This filter is typically used in normal single-particle tracking (SPT) scenarios, 
but should be disabled if particles might be truly static, such as in very viscous media.

'''

'''
THINGS TO DO:

apply_consecutive_stop_filter:
- Set to 1 to apply the filter (remove stuck particles), or 0 to disable it.

stop_NMs:
- Number of consecutive frames with zero displacement required to classify a particle as ŌĆ£stuck.ŌĆØ  
- Typical value: 5 frames at ~ 40ŌĆ»fps.
  * For higher frameŌĆærates, raise the value proportionally (e.g., 8 at 60ŌĆ»fps).
  
'''

##################################
apply_consecutive_stop_filter = 1  
stop_NMs = 5
##################################

import numpy as np

if apply_consecutive_stop_filter:
    displacements = t1.groupby('particle')[['x', 'y']].agg(lambda df: np.sqrt((df.diff()**2).sum())).sum(axis=1)
    valid_particles = displacements[displacements >= stop_NMs].index 
    t2 = t1[t1['particle'].isin(valid_particles)]

    t2_filtered = t2.copy()
    t2_filtered['dx'] = t2_filtered.groupby('particle')['x'].diff().fillna(0)
    t2_filtered['dy'] = t2_filtered.groupby('particle')['y'].diff().fillna(0)
    t2_filtered['movement'] = np.sqrt(t2_filtered['dx']**2 + t2_filtered['dy']**2)

    particles_to_remove = set()
    for particle, group in t2_filtered.groupby('particle'):
        stopped_frames = (group['movement'] == 0).astype(int)
        consecutive_stopped = stopped_frames.rolling(window=5).sum().fillna(0)
        if (consecutive_stopped >= stop_NMs).any():
            particles_to_remove.add(particle)

    t2_filtered = t2_filtered[~t2_filtered['particle'].isin(particles_to_remove)]
else:
    t2_filtered = t1.copy()
```

```python
'''
EXPLANATION OF SCRIPT 9: DUPLICATE TRAJECTORY FILTERING

The goal of this script is to ensure that tracked trajectories correspond
to unique particles without duplication. After a track of a nanomotor ends,
no new track can start within `disappear_radius` pixels or
`disappear_frames` frames of its last position.

'''

'''
THINGS TO DO:

ŌåÆ Adjust these parameters as needed before running the code:

disappear_radius:  
- Pixel radius around the particleŌĆÖs last position (x,y) where new tracks are not accepted.  
- Larger radius = stricter duplication control.
  * Recomendation: set it Ōł╝ŌĆ»2ŌĆ»├ŚŌĆ»`max_diameter` chosen in SCRIPTŌĆ»4.

disappear_frames:  
- Number of frames this "protected area" remains active after a track ends.   
  * Recomendation:ŌĆ»(fpsŌĆ»/ŌĆ»2), or higher.

'''

##############################
disappear_radius = 100
disappear_frames = 60     # must be an integer number (that's why we use int())
##############################

import numpy as np
last_positions = t2_filtered.groupby('particle').last()[['x', 'y', 'frame']]
restricted_zones = []
for _, row in last_positions.iterrows():
    x_last, y_last, frame_last = row['x'], row['y'], int(row['frame'])  

    for future_frame in range(frame_last + 1, frame_last + disappear_frames): # "30" frames
        
        restricted_zones.append((future_frame, x_last, y_last))
def is_in_restricted_zone(frame, x, y):
    for f, x_r, y_r in restricted_zones:

        if f == frame and np.sqrt((x - x_r)**2 + (y - y_r)**2) <= disappear_radius: 
            
            return True
    return False
first_appearances = t2_filtered.groupby('particle').first()[['x', 'y', 'frame']]
valid_particles = first_appearances[~first_appearances.apply(lambda row: is_in_restricted_zone(int(row['frame']), row['x'], row['y']), axis=1)].index
t3_filtered = t2_filtered[t2_filtered['particle'].isin(valid_particles)]
print(f'Number of particles before: {t2_filtered["particle"].nunique()}')
print(f'Number of particles after applying the constraint: {t3_filtered["particle"].nunique()}')
tp.plot_traj(t3_filtered)
```

```python
"""
This script filters out weird or noisy MSD trajectories from particle data.

It checks each trajectory for:
1. Too many ups and downs (local peaks).
2. Big jumps between high and low values.
3. A drop of at least drop_threshold units at any later time point (even if not consecutive).

If a trajectory looks too irregular, it's removed. 
The rest are plotted so we only see the smoother, more reliable ones.
"""

import numpy as np
import scipy.signal
import matplotlib.pyplot as plt


######################
max_peaks = 2
min_diff = 1
drop_threshold=0.1
######################

# Function to detect irregular MSD trajectories
def detect_irregular_msd(msd_data, max_peaks, min_diff, drop_threshold=1.0):
    irregular_particles = []

    for particle in msd_data.columns:
        msd_values = msd_data[particle].values

        # Detect local maxima and minima
        max_idx = scipy.signal.argrelextrema(msd_values, np.greater)[0]
        min_idx = scipy.signal.argrelextrema(msd_values, np.less)[0]
        num_peaks = len(max_idx) + len(min_idx)

        # Difference between average maxima and minima
        if len(max_idx) > 0 and len(min_idx) > 0:
            max_min_diff = np.mean(msd_values[max_idx]) - np.mean(msd_values[min_idx])
        else:
            max_min_diff = 0

        # Check for any future drop of at least drop_threshold
        has_big_drop = False
        for i in range(len(msd_values)):
            future_vals = msd_values[i+1:]
            if any(msd_values[i] - future_val >= drop_threshold for future_val in future_vals):
                has_big_drop = True
                break

        # Mark as irregular if any condition is met
        if num_peaks > max_peaks or max_min_diff > min_diff or has_big_drop:
            irregular_particles.append(particle)

    return irregular_particles

# Example usage (assuming im_filtered is a DataFrame with MSD data)
irregular_particles = detect_irregular_msd(im_filtered, max_peaks, min_diff, drop_threshold)
filtered_msd = im_filtered.drop(columns=irregular_particles)

tX_filtered = t4_filtered[t4_filtered['particle'].isin(filtered_msd.columns)]  # Filtrar part├Łculas limpias


# Plot filtered MSD trajectories
fig, ax = plt.subplots()
plt.title('Mean Square Displacement (Filtered)')
ax.plot(filtered_msd.index, filtered_msd, 'k-', alpha=0.1)
ax.set(ylabel=r'$\langle \Delta r^2 \rangle$ [$\mu$m$^2$]', xlabel='lag time $t$ [s]')
plt.show()
print("Number of particles before removal:", tX_filtered['particle'].nunique())
```

```python
# Here we calculate the polar map and save the data in an Excel file

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df__copy = t5_filtered.copy()

df__copy = df__copy.reset_index(drop=True)


# THINGS TO CHANGE
video_width = 2048
video_height = 2048
frame_interval = 750 # in this case, 6*25(fps) = 6s 
partitions = 24 # NUMBER OF GROUPS


angles = []
radii = []

for pid, group in df__copy.groupby('particle'):
    group = group.sort_values(by='frame')
    frame_start = group['frame'].min()
    frame_end = frame_start + frame_interval

    pos_start = group[group['frame'] == frame_start]
    pos_end = group[group['frame'] == frame_end]

    if not pos_start.empty and not pos_end.empty:
        dx = pos_end['x'].values[0] - pos_start['x'].values[0]
        dy = (video_height - pos_end['y'].values[0]) - (video_height - pos_start['y'].values[0])
        r = np.sqrt(dx**2 + dy**2)
        theta = np.arctan2(dy, dx)
        angles.append(theta)
        radii.append(r)


angles = np.array(angles)
radii = np.array(radii)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, polar=True)

ax.set_theta_zero_location("E")  
ax.set_theta_direction(1)   
ax.set_thetagrids(np.arange(0, 360, 15))  

bins = np.linspace(-np.pi, np.pi, partitions+1)  
counts, _ = np.histogram(angles, bins=bins)

bin_centers = (bins[:-1] + bins[1:]) / 2

width = (2 * np.pi) / (partitions)
bars = ax.bar(bin_centers, counts, width=width, bottom=0.0,
              alpha=0.7, color='teal', edgecolor='black')

ax.set_title('Polar Map', va='bottom')
plt.show()

####################################################################################

import pandas as pd
import numpy as np

bins_deg = np.linspace(0, 360, partitions+1)  
counts, _ = np.histogram(np.rad2deg(angles) % 360, bins=bins_deg)  

ranges = [f"{int(bins_deg[i])}┬░ŌĆō{int(bins_deg[i+1])}┬░" for i in range(len(bins_deg) - 1)]

df_hist = pd.DataFrame({
    "Angle (range)": ranges,
    "Number of particles": counts
})

# Guardar en Excel
df_hist.to_excel("POLAR_1.xlsx", index=False) # NAME the Excel File
print("Data saved")
```
